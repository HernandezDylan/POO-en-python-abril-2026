Conceptos clave que aplicamos:
Clase (class): Es el "molde" o plantilla para crear perros.

Constructor (__init__): Es la función especial que se ejecuta automáticamente cuando creas un objeto. Sirve para asignar el Nombre y la Raza desde el inicio.

Self: Es una palabra clave que le dice a Python: "estoy hablando de los datos de este perro en específico".

Instancia (Objetos): pitbull y chihuahua son dos "hijos" de la clase Animal, cada uno con su propia personalidad (datos).

Nota para principiantes: En Python, los nombres de las clases suelen empezar con mayúscula (Animal) y las funciones con minúscula (ladrar). ¡Es una buena práctica que te hará ver como un profesional!

```python
# Definimos la clase
class Animal:
    # El constructor __init__ inicializa los atributos al crear el objeto
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función (método) para ladrar
    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau, guau!🐶")

    # Función (método) para correr
    def correr(self):
        print(f"{self.nombre} (raza {self.raza}) está corriendo velozmente.🐕")

    def Morder(self):
        print(f"{self.nombre} está mordiendo a la vecina  🤦‍♀️ ")

# --- Creación de los objetos (Instanciación) ---

# Creamos el objeto 'pitbull' con datos específicos
pitbull = Animal("Rocco", "Pitbull")

# Creamos el objeto 'chihuahua' con datos específicos
chihuahua = Animal("Chispa", "Chihuahua")
# creamos el objeto perro callejero con datos
perrocallejero = Animal("solovino", "chafa")

# --- Uso de los objetos ---

# Ejecución de acciones para el Pitbull
print("--- Datos del primer perro ---")
pitbull.ladrar()
pitbull.correr()
pitbull.Morder()

print("\n--- Datos del segundo perro ---")
# Ejecución de acciones para el Chihuahua
chihuahua.ladrar()
chihuahua.correr()
chihuahua.Morder()

print("\n--- Datos del perro callejero ---")
# Ejecución de acciones para el perro callejero
perrocallejero.ladrar()
perrocallejero.correr()
perrocallejero.Morder()

print("Programa realizado por Hernadez Dylan NL=24")
```
